```python
import numpy as np
import pandas as pd
import matplotlib.image as img
import matplotlib.pyplot as plt
```

```python
import os

file_direc=os.listdir(r"C:\Users\haider\Desktop\New folder (3)\cifar_images")
print(file_direc)
```

```python
list_label=[]

for file in file_direc:
    if file[:7]=="label_0":
        list_label.append(0)
    elif file[:7]=="label_1":
        list_label.append(1)
    elif file[:7]=="label_2":
        list_label.append(2)
    elif file[:7]=="label_3":
        list_label.append(3)
    elif file[:7]=="label_4":
        list_label.append(4)
    elif file[:7]=="label_5":
        list_label.append(5)
    elif file[:7]=="label_6":
        list_label.append(6)
    elif file[:7]=="label_7":
        list_label.append(7)
    elif file[:7]=="label_8":
        list_label.append(8)
    elif file[:7]=="label_9":
        list_label.append(9)

print(list_label)
```

```python
label_data=np.array(list_label)
label_data
```

```python
print(file_direc)
```

```python
from PIL import Image

path_image="C:\\Users\\haider\\Desktop\\New folder (3)\\cifar_images\\"

data=[]

for files in file_direc:
    image=Image.open(path_image+files)
    image_array=np.array(image)
    data.append(image_array)

data_object=np.array(data)
print(data_object)
```

```python
print(type(label_data),type(data_object))
print(label_data.shape, data_object.shape)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(data_object,label_data, test_size=.2, random_state=42)
print(data_object.shape,label_data.shape,train_x.shape,test_x.shape,train_y.shape,test_y.shape)
```

```python
train_x_set=train_x/255
test_x_set=test_x/255
```

```python
import tensorflow as tf
# from tensorflow.keras.applications import ResNet50

# convolutional_bade=ResNet50(weights="imagenet", include_top=False, input_shape=(256,256,3))
# convolutional_bade.trainable=False
```

```python
model=tf.keras.Sequential([
    
    tf.keras.layers.Conv2D(32, kernel_size=(3,3), activation="relu", input_shape=(32,32,3)),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    tf.keras.layers.Conv2D(64, kernel_size=(3,3), activation="relu"),
    tf.keras.layers.MaxPooling2D(pool_size=(2,2)),
    
    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(64, activation="relu"),
    tf.keras.layers.Dropout(0.5),

    tf.keras.layers.Dense(10, activation="softmax")
])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
history=model.fit(train_x_set,train_y, validation_split=.1, epochs=10)
```

```python
loss, accuracy =model.evaluate(test_x_set,test_y)
```

```python
pre=model.predict(test_x_set)
pre[20]
```

```python
plt.plot(history.history["accuracy"],"r", label="accuracy")
plt.plot(history.history["val_accuracy"],"b", label="val_accuracy")
plt.legend()
```
